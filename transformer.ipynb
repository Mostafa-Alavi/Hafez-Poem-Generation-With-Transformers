# Tokenizing & Embedding

```python
from hazm import word_tokenize
from gensim.models import Word2Vec
import numpy as np
import torch

# Tokenize the Persian text
odd_lines_tokens = [word_tokenize(line) for line in odd_lines]
even_lines_tokens = [word_tokenize(line) for line in even_lines]

print(f"First 20 tokens: {even_lines_tokens[:20]}")
print(f"Total number of tokens: {len(even_lines_tokens)}")

# Create sentences (each token as a single word sentence for poetry)
sentences = odd_lines_tokens + even_lines_tokens
print(sentences)
# # Train Word2Vec model
word2vec_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
```

```python

max_len = max(len(line) for line in sentences)
print(f"Maximum tokens in a sentence: {max_len}")

# Create a list for padded sequences
padded_sequences = []

# Pad sequences with '<pad>' token
for sequence in sentences:
    # Get word vector for each token
    sequence_vectors = [word2vec_model.wv[token] for token in sequence]
    
    # Pad with zeros if needed
    padding_length = max_len - len(sequence)
    if padding_length > 0:
        # Add '<pad>' token vector (zeros) for padding
        # First add '<pad>' to word2vec model if not already present
        print(f"word2vec_model.wv : {'<pad>' not in word2vec_model.wv }")
        if '<pad>' not in word2vec_model.wv:
            word2vec_model.build_vocab([['<pad>']], update=True)
        padding_vector = word2vec_model.wv['<pad>']
        sequence_vectors.extend([padding_vector] * padding_length)
    
    padded_sequences.append(sequence_vectors)

# Convert to tensor
padded_tensor = torch.tensor(padded_sequences)
print(f"Shape of padded tensor: {padded_tensor.shape}")
```

```python
train_split_percent = 0.8
X_data_embeddings = padded_tensor[:int(len(padded_tensor)//2), :, :]
y_data_embeddings = padded_tensor[int(len(padded_tensor)//2):, :, :]
print(f"X_data_embeddings.shape: {X_data_embeddings.shape}")

X_train_embeddings = X_data_embeddings[:int(train_split_percent*len(X_data_embeddings))]
X_test_embeddings = X_data_embeddings[int(train_split_percent*len(X_data_embeddings)):]

y_train_embeddings = y_data_embeddings[:int(train_split_percent*len(y_data_embeddings))]
y_test_embeddings = y_data_embeddings[int(train_split_percent*len(y_data_embeddings)):]

X_train_embeddings.shape, y_test_embeddings.shape
```

```python

def generate_x_y_batches(data:torch.tensor, context_len:int , step:int , batch_size):
    x_batches = []
    y_batches = []


    for i in range(0, len(data)-context_len, step):
        x_batches.append(data[i:i+context_len])
        y_batches.append(data[i+1: i+1+context_len])
    
    num_sequences = len(x_batches)
    num_complete_batches = (num_sequences // batch_size) * batch_size

    x_batches = x_batches[:num_complete_batches]
    y_batches = y_batches[:num_complete_batches]
    
    x_batch_tensor = torch.stack(x_batches)
    y_batch_tensor = torch.stack(y_batches)
    
    x_batch_tensor = x_batch_tensor.view(-1, batch_size, context_len, 100)
    y_batch_tensor = y_batch_tensor.view(-1, batch_size, context_len, 100)

    return x_batch_tensor, y_batch_tensor


Xb, yb = generate_x_y_batches(data = train_data_embeddings, context_len = 20, step =10 , batch_size = 50)
print("Xb shape:", Xb.shape)
print("yb shape:", yb.shape)


# Print actual values for comparison
print("X[0,0,1,:5]:", Xb[0,0,1,:5])
print("y[0,0,0,:5]:", yb[0,0,0,:5])

# First check if shapes match
assert Xb.shape == yb.shape, "Batch shapes don't match"

# Check if elements are equal using torch.allclose() which handles floating point differences
# Note: Using allclose instead of exact equality because we're dealing with floats
assert torch.allclose(Xb[0,0,1,:5], yb[0,0,0,:5], rtol=1e-5), "Sequential elements don't match"
```

# Model

```python
from torch import nn

class HeadAttention(nn.Module):

    def __init__(self, embedding_dim:int , head_size:int, context_len:int , dropout :float):
        super().__init__()
        self.head_size = head_size
        self.embedding_dim = embedding_dim
        self.query = nn.Linear(self.embedding_dim, self.head_size, bias=False)
        self.key = nn.Linear(self.embedding_dim, self.head_size, bias=False)
        self.value = nn.Linear(self.embedding_dim, self.head_size, bias=False)
        self.softmax = nn.Softmax(dim=-1)
        self.dropput = nn.Dropout(dropout)
        self.register_buffer('tril', torch.tril(torch.ones(context_len, context_len)))
        
    
    
    def forward(self, x):
        # x shape: (batch_size, context_len, embedding_dim)
        # Project input x into query, key, and value
        queries = self.query(x)
        keys = self.key(x)
        values = self.value(x)

        # Compute attention scores (use scaled dot-product attention)
        # queries shape: (batch_size, context_len, head_size)
        # keys shape: (batch_size, context_len, head_size)
        # Scale dot-product by sqrt(head_size)
        scores = (queries @ keys.transpose(-2, -1)) / (self.head_size ** 0.5)

        scores = scores.masked_fill(self.tril== 0, float('-inf')) # (B, T, T)
        scores = self.softmax(scores) # (B, T, T)
        scores = self.dropput(scores)
        attention = scores @ values  # Apply softmax to get attention weights

        
        return attention



class FeedForward(nn.Module):
    def __init__(self, embedding_dim:int, dropout:float):
        super().__init__()
        self.embedding_dim = embedding_dim
        self.hidden_dim = 4*embedding_dim
        self.dropout = nn.Dropout(dropout)
        self.relu = nn.ReLU()
        self.linear1 = nn.Linear(self.embedding_dim, self.hidden_dim)
        self.linear2 = nn.Linear(self.hidden_dim, self.embedding_dim)
    
    def forward(self, x):
        x = self.linear1(x)
        x = self.relu(x)
        x = self.linear2(x)
        return self.dropout(x)



class MultiHeadAttention(nn.Module):
    def __init__(self, embedding_dim:int, context_len:int,  head_size:int, num_heads:int, dropout :float):
        super().__init__()
        self.head_size = head_size
        self.num_heads = num_heads
        self.embedding_dim = embedding_dim
        self.heads = nn.ModuleList([HeadAttention(embedding_dim, head_size, context_len,  dropout) for _ in range(num_heads)])
        self.out = nn.Linear(self.num_heads * self.head_size, self.embedding_dim)

        self.dropout = nn.Dropout(dropout)
    
    def forward(self, x):
        heads_ouput = [head(x) for head in self.heads]
        heads_ouput_tensor = torch.stack(heads_ouput, dim=3)
        heads_ouput_tensor = heads_ouput_tensor.view(x.shape[0], x.shape[1], self.num_heads * self.head_size)

        return self.dropout(self.out(heads_ouput_tensor))


class TransformerBlock(nn.Module):
    def __init__(self, embedding_dim, context_len ,  head_size, num_heads, dropout ):
        super().__init__()
        head_size = embedding_dim//num_heads
        self.multi_head_attention = MultiHeadAttention(embedding_dim=embedding_dim, context_len=context_len, head_size=head_size, num_heads=head_size, dropout=dropout)
        self.feed_forward = FeedForward(embedding_dim, dropout)
        self.layer_norm1 = nn.LayerNorm(embedding_dim)
        self.layer_norm2 = nn.LayerNorm(embedding_dim)
    
    def forward(self, x):
        attention_output = self.multi_head_attention(x)
        x = self.layer_norm1(x + attention_output)
        feed_forward_output = self.feed_forward(x)
        return self.layer_norm2(x + feed_forward_output)
    


class GPTLanguageModel(nn.Module):
    def __init__(self, embedding_dim:int, context_len:int, head_size:int, num_heads:int, num_transformer_blocks:int, dropout:float):
        super().__init__()
        self.embedding_dim = embedding_dim
        self.context_len = context_len
        self.head_size = head_size
        self.num_heads = num_heads
        self.num_transformer_blocks = num_transformer_blocks
        self.dropout = dropout
        self.transformer_blocks = nn.ModuleList([TransformerBlock(embedding_dim, context_len , head_size, num_heads, dropout) for _ in range(num_transformer_blocks)])
        self.layernorm_final = nn.LayerNorm(embedding_dim) # final layer norm
        self.out = nn.Linear(embedding_dim, embedding_dim)

        self.position_embedding_table = nn.Embedding(context_len, embedding_dim)

    
    def forward(self, x):
        x = x + self.position_embedding_table(torch.arange(self.context_len, device=x.device))
        
        for transformer_block in self.transformer_blocks:
            x = transformer_block(x)
        
        x = self.layernorm_final(x) # (B,T,C)
        logits = self.out(x) # (B,T,vocab_size)
        return logits


@torch.no_grad()    
def generate(model, x, max_tokens_len:int):
    # x , y shape: ( batch_size, context_len, embedding_dim)
    context_len = x.shape[1]
    
    model.eval()

    original_text = []
    for i  in range(context_len):  # Taking first batch, first sequence, first token, embedding_vector
        # Find most similar word to the embedding
        similar_words = word2vec_model.wv.similar_by_vector(x[0,i,:].cpu().detach().numpy())
        if similar_words:
            original_text.append(similar_words[0][0])  # Get the most similar word
            
    # Get the most similar words for generated embeddings
    generated_text = []
    for _ in range(max_tokens_len):
        
        y = model(x) 

        similar_words = word2vec_model.wv.similar_by_vector(y[0,-1,:].cpu().detach().numpy())
        print(f"Given text : {' '.join(original_text + generated_text)}:")
        print(f"{similar_words[0][0]}")
        if similar_words:
            generated_text.append(similar_words[0][0])  # Get the most similar word
        

        x = torch.cat([x, y[:,-1,:].unsqueeze(1)], dim=1)  # Add the predicted token to the context
        x = x[:, -context_len:, :]  # Keep only the last context_len tokens
        

        
    return y
    


# @torch.no_grad()    
# def generate(model, x, max_tokens_len:int):
#     # x , y shape: (num_batchees, batch_size, context_len, embedding_dim)
#     context_len = x.shape[2]
    
#     model.eval()

#     original_text = []
#     for i  in range(x.shape[2]):  # Taking first batch, first sequence, first token, embedding_vector
#         # Find most similar word to the embedding
#         similar_words = word2vec_model.wv.similar_by_vector(x[0,1,i,:].cpu().detach().numpy())
#         if similar_words:
#             original_text.append(similar_words[0][0])  # Get the most similar word
            
#     # Get the most similar words for generated embeddings
#     generated_text = []
 
#     y = model(x) 

#     for i in range(x.shape[2]):
#         similar_words = word2vec_model.wv.similar_by_vector(y[0,1,i,:].cpu().detach().numpy())
#         if similar_words:
#             generated_text.append(similar_words[0][0])  # Get the most similar word
        
#     print(f"Given text : {' '.join(original_text)}:")
#     print(f"Generated : {' '.join(generated_text)}")
           

        
#     return y
    
```

# Training

```python
from tqdm import tqdm

num_train_epochs = 10
batch_size = 512
step_token = 10
context_len = 15
embedding_dim = 100
head_size=8
num_heads = 8
num_transformer_blocks = 3
dropout = 0.2
device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = GPTLanguageModel(embedding_dim=embedding_dim, context_len=context_len, num_transformer_blocks = num_transformer_blocks , head_size=head_size, num_heads=num_heads, dropout=dropout)
model.to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=0.001)


Xb_train, yb_train = generate_x_y_batches(X_embeddings=X_train_embeddings, y_embeddings=y_train_embeddings, context_len = context_len , batch_size = batch_size)

Xb_test, yb_test = generate_x_y_batches(X_embeddings=X_test_embeddings, y_embeddings=y_test_embeddings, context_len = context_len , batch_size = batch_size)

# Move to device and ensure gradients are maintained
Xb_train, yb_train = Xb_train.to(device), yb_train.to(device)
Xb_test, yb_test = Xb_test.to(device), yb_test.to(device)

print(f"Xb_train.shape: {Xb_train.shape}")
# Enable gradient computation
Xb_train.requires_grad_(True)
Xb_test.requires_grad_(True)

def cosine_similarity_loss(pred, target, padding_mask=None):
    # pred shape: [batch_size, seq_len, embedding_dim]
    # target shape: [batch_size, seq_len, embedding_dim]
    # padding_mask shape: [batch_size, seq_len]
    
    # Normalize predictions and targets
    pred_norm = torch.nn.functional.normalize(pred, dim=-1)
    target_norm = torch.nn.functional.normalize(target, dim=-1)
    
    # Calculate cosine similarity
    similarity = torch.sum(pred_norm * target_norm, dim=-1)  # [batch_size, seq_len]
    
    # If padding mask is provided, apply it
    if padding_mask is not None:
        # Convert padding mask to float and expand to match similarity shape
        mask = padding_mask.float()
        # Apply mask and compute mean only over non-padded positions
        # Sum across both batch and sequence dimensions for masked positions
        return 1 - (torch.sum(similarity * mask) / torch.sum(mask).clamp(min=1))
    else:
        return 1 - torch.mean(similarity)


for epoch in range(num_train_epochs):
    model.train()
    avg_loss=0
    for Xb, yb in tqdm(zip(Xb_train, yb_train)):
        optimizer.zero_grad()

        # Forward pass
        # Create padding mask for the entire batch
        # Xb shape: [batch_size, seq_len]
        pad_token = torch.tensor(word2vec_model.wv['<pad>'], device=device)
        # print(f"pad_token.shape: {pad_token.shape}")
        # Compare the entire embedding vectors and take any() to get a single boolean per position
        padding_mask = 1-(torch.all(Xb == pad_token, dim=-1).float())  # [batch_size, seq_len] 
        # print(f"padding_mask.shape: {padding_mask.shape}")
        # print(f"padding_mask: {padding_mask}")
        # Forward pass
        yb_train_prediction = model(Xb)  # [batch_size, seq_len, embedding_dim]
        
        # Calculate loss with padding mask
        loss = cosine_similarity_loss(
            yb_train_prediction,  # [batch_size, seq_len, embedding_dim]
            yb,                  # [batch_size, seq_len, embedding_dim]
            padding_mask        # [batch_size, seq_len]
        )
        loss.backward()
        optimizer.step()
        # print(f"loss: {loss.item()}")
        avg_loss+=loss.item()
        
    
    avg_loss/=len(Xb_train)
    print(f" epoch : {epoch} , Train Loss: {avg_loss}")
    # Generation of test data
    generated_text = generate(model, Xb_test[1], max_tokens_len=context_len)
    # print(f"generated_test : {generated_test.shape}")



```
